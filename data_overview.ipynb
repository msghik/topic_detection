```python
import pandas as pd

data = pd.read_csv("/home/mohammad/Downloads/Telegram Desktop/output.csv")

data
```

```python
# Let's drop the rows that has NaN values

data = data.dropna()
data
```

```python
data.button.value_counts().to_list()
```

```python
data.button.value_counts()

# create a visualization of the count of each categorise in the button column

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

matplotlib.rcParams['figure.figsize'] = (20, 10)
sns.set(style="whitegrid")
sns.set_palette("pastel")
sns.countplot(x='button', data=data, order=data['button'].value_counts().index)
plt.title("Count of each category in the button column")
plt.xlabel("Button")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```
